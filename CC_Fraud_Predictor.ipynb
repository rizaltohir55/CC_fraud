```python
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score, precision_score, recall_score, roc_curve, auc

# Fungsi untuk menghitung dan menampilkan metrik
def evaluate_model(y_test, predictions, model_name):
    # Menghitung AUC
    fpr, tpr, _ = roc_curve(y_test, predictions)
    auc_score = auc(fpr, tpr)

    # Menghitung akurasi, presisi, recall, dan F1 score
    accuracy = accuracy_score(y_test, predictions)
    f1 = f1_score(y_test, predictions)
    precision = precision_score(y_test, predictions)
    recall = recall_score(y_test, predictions)

    # Menampilkan hasil evaluasi
    print(f"Model: {model_name}")
    print(f"AUC: {auc_score:.4f}")
    print(f"Akurasi: {accuracy*100:.2f}%")
    print(f"F1 Score: {f1:.4f}")
    print(f"Presisi: {precision:.4f}")
    print(f"Recall: {recall:.4f}")

    # Menampilkan ROC Curve
    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {auc_score:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {model_name}')
    plt.legend(loc='lower right')
    plt.show()

    return auc_score, accuracy, f1, precision, recall
```

```python
# Evaluasi untuk Random Forest
rf_auc, rf_accuracy, rf_f1, rf_precision, rf_recall = evaluate_model(y_test, rf_pred, "Random Forest")

# Evaluasi untuk XGBoost
xgb_auc, xgb_accuracy, xgb_f1, xgb_precision, xgb_recall = evaluate_model(y_test, xgb_pred, "XGBoost")

# Evaluasi untuk LightGBM
lgb_auc, lgb_accuracy, lgb_f1, lgb_precision, lgb_recall = evaluate_model(y_test, lgb_pred, "LightGBM")

# Evaluasi untuk Gradient Boosting
gb_auc, gb_accuracy, gb_f1, gb_precision, gb_recall = evaluate_model(y_test, gb_pred, "Gradient Boosting")
```

```python
import plotly.graph_objects as go

# Membuat Data untuk Visualisasi
models = ["Random Forest", "XGBoost", "LightGBM", "Gradient Boosting"]
accuracy_scores = [rf_accuracy, xgb_accuracy, lgb_accuracy, gb_accuracy]
f1_scores = [rf_f1, xgb_f1, lgb_f1, gb_f1]
precision_scores = [rf_precision, xgb_precision, lgb_precision, gb_precision]
recall_scores = [rf_recall, xgb_recall, lgb_recall, gb_recall]

# Visualisasi Akurasi (Bar Chart)
fig_accuracy = go.Figure(data=[go.Bar(x=models, y=accuracy_scores, name="Akurasi", marker_color='blue')])
fig_accuracy.update_layout(title="Perbandingan Akurasi Model", xaxis_title="Model", yaxis_title="Akurasi (%)")
fig_accuracy.show()

# Visualisasi F1 Score (Bar Chart)
fig_f1 = go.Figure(data=[go.Bar(x=models, y=f1_scores, name="F1 Score", marker_color='green')])
fig_f1.update_layout(title="Perbandingan F1 Score Model", xaxis_title="Model", yaxis_title="F1 Score")
fig_f1.show()

# Visualisasi Presisi (Bar Chart)
fig_precision = go.Figure(data=[go.Bar(x=models, y=precision_scores, name="Presisi", marker_color='orange')])
fig_precision.update_layout(title="Perbandingan Presisi Model", xaxis_title="Model", yaxis_title="Presisi")
fig_precision.show()

# Visualisasi Recall (Bar Chart)
fig_recall = go.Figure(data=[go.Bar(x=models, y=recall_scores, name="Recall", marker_color='red')])
fig_recall.update_layout(title="Perbandingan Recall Model", xaxis_title="Model", yaxis_title="Recall")
fig_recall.show()
```

Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Tentukan parameter grid yang akan diuji
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Membuat objek RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# Menggunakan GridSearchCV untuk melakukan hyperparameter tuning
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# Fitting GridSearchCV
grid_search_rf.fit(X_train, y_train)

# Menampilkan parameter terbaik yang ditemukan
print(f"Best Hyperparameters for Random Forest: {grid_search_rf.best_params_}")

# Evaluasi model dengan parameter terbaik
best_rf_model = grid_search_rf.best_estimator_
rf_pred_best = best_rf_model.predict(X_test)

# Menghitung akurasi dan metrik lainnya setelah tuning
rf_accuracy_best = accuracy_score(y_test, rf_pred_best) * 100
print(f"Akurasi Random Forest Setelah Tuning: {rf_accuracy_best:.2f}%")
```
